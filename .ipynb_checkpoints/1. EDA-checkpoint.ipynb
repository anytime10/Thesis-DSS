```python
# relevant imports:
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import psynlig as pl
import missingno as mmsno
import math
```

```python
# loading the merged csv dataset:
data_22 = pd.read_csv('italy_2022.csv')
print(data_22.shape)
print(data_22.columns)
```

```python
# renaming columns with unclear names:
data_22.rename(columns={'CNTSCHID': 'schoolID','CNTSTUID': 'studentID','ST004D01T': 'Gender','SC001Q01TA': 'municipality_size', 
                        'SC211Q03JA': 'disadvantaged_students','SC037Q02TA': 'external_evaluation','SC187Q04WA': 'math_grouping',
                          }, inplace=True)
print(data_22.columns)
```

```python
# compute the proportion of missing values in each column in the dataset:
missing_values = data_22.isnull().mean()
missing_values = missing_values[missing_values > 0]
missing_values = missing_values.sort_values(ascending=False)
print(missing_values)
```

```python
# checking for duplicated rows:
duplicated_rows = data_22.duplicated().sum()
print(duplicated_rows)
```

# Missing data analysis

```python
'''function that filters out columns below a certain threshold of missing values - except for some columns that are considered
 important - and returns the remaining columns:'''

important_cols = ['PROBSELF','SCHSUST','PROBSCRI', 'disadvantaged_students', 'SMRATIO', 'STAFFSHORT',
                  'CLSIZE']
def filter_missing_values(data, threshold):
    missing_values = data.isnull().mean()
    missing_values = missing_values[missing_values > threshold]
    missing_values = missing_values.sort_values(ascending=False)
    cols_to_drop = missing_values[missing_values > threshold].index
    cols_to_drop = cols_to_drop.difference(important_cols)
    data = data.drop(cols_to_drop, axis=1)
    return data
df_filtered = filter_missing_values(data_22, 0.15)
df_filtered.shape
```

```python
# Split the dataframe into two parts to visualize more easily:
df_1 = df_filtered.iloc[:, :int(df_filtered.shape[1] / 2)]
df_2 = df_filtered.iloc[:, int(df_filtered.shape[1] / 2):]
# Set up a figure with two subplots 
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(20, 20))  # Adjust the figsize as necessary
    
mmsno.bar(df_1, ax=axs[0])
axs[0].set_title('Missing Data for First Half of Columns')
    
mmsno.bar(df_2, ax=axs[1])
axs[1].set_title('Missing Data for Second Half of Columns')
 
plt.tight_layout()
plt.show()
```

```python
# missing data matrix:
mmsno.matrix(df_filtered)
```

```python
# plotting the missing data heatmap:
mmsno.heatmap(df_filtered)
```

```python
# plotting spearman correlation matrix:
df_heatmap = df_filtered.drop(columns = ['schoolID','studentID','PV1MATH'])
continuos_features = ['SMRATIO','PROPSUPP','PROATCE','disadvantaged_students','SCHAUTO', 'GRADE', 'STRATIO', 'RATCMP1','PROBSCRI',
                      'HOMEPOS','ESCS','STUBEHA','TEACHBEHA','INSTLEAD','DIGPREP', 'SCHSUST','PROBSELF','EDUSHORT', 'FAMSUP',
                       'STAFFSHORT','SCHSIZE','TEACHSUP']
df_heatmap = df_heatmap[continuos_features]
corr = df_heatmap.corr(method='spearman')
plt.figure(figsize=(20, 20))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spearman Correlation Matrix')
plt.show()
```
